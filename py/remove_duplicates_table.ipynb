```python
import numpy as np
from tables import *
import csv
from tqdm import tqdm_notebook as tqdm
```

```python
class SimPairs(IsDescription):
    Id_Art1 = UInt32Col()
    Id_Art2 = UInt32Col()
    Sim = Float16Col()
```

```python
data_path = "/mnt/disks/sdc/data/"
```

```python
years = np.arange(1898,2020)
```

```python
h5file_light = open_file(data_path + 'similarity_light.h5', mode="w", title="similarity")
group = h5file_light.create_group("/", "light")
for year in years:
    h5file_light.create_table(group, str(year), SimPairs)
```

```python
h5file_sim = open_file(data_path+"similarity.h5", mode="r", title="similarity")
h5file_light = open_file(data_path + 'similarity_light.h5', mode="a", title="similarity")
nb_lines = 0
for y in tqdm(range(len(years))):
    year = years[len(years)-y-1]
    table_light = h5file_light.get_node('/light/{}'.format(str(year)))
    row = table_light.row
    pairs = set()
    for i in range(10):
        table_sim = h5file_sim.get_node('/{}/{}'.format(str(year),i))
        for j in range(len(table_sim)):
            line = table_sim[j]
            tup_pair = (line['Id_Art1'],line['Id_Art2'])
            if tup_pair not in pairs:
                row['Id_Art1'] = tup_pair[0]
                row['Id_Art2'] = tup_pair[1]
                row['Sim'] = line['Sim']
                row.append()
                pairs.add(tup_pair)
    table_light.flush()
h5file_sim.close()
h5file_light.close()
```
