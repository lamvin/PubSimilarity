```python
DictAbs = {}
for i in tqdm(range(nb_l_data)):
    line = table_data[i]
    Art_Id = line['Id_Art']
    if Art_Id in DictAbs.keys():
        DictAbs[Art_Id].append(i)
    else:
        DictAbs[Art_Id] = [i]
abs_cits = DictAbs.keys()
```

```python
def read_ID(idx,table):
    rows = table[idx]['Abstract']
    rows = [r.decode("utf-8") for r in rows]
    return ' '.join(rows).split(' ')
```

```python
for i in tqdm(range(nb_l_cits)):
    line_cits = table_cits[i]
    citant = line_cits['citant']
    cites = np.fromstring(line_cits['cites'],dtype=np.int32)
    comps = []
    for cite in cites:
        if cite in abs_cits:
            comps.append(cite)
    if citant in abs_cits:
        comps.append(citant)
    corpus = []
    for Art_ID in comps:
        corpus.append(read_ID(DictAbs[Art_ID],table_data))
 
    dct = Dictionary([[line for line in doc] for doc in corpus])
    bow = [dict_data.doc2bow(vec_i) for vec_i in corpus]
    model = TfidfModel(bow)  
    tfidf = model[bow]
    index = similarities.SparseMatrixSimilarity(tfidf,num_features=len(dct)) #check if num ft has impact on time
    sim = index[tfidf]
    
```
